Supongamos que tenemos el registro de las ventas del primer semestre del 2022 de una empresa de ventas online en el que figuran: nombre del vendedor, rubro de la venta, sede de distribución, importe y fecha de la operación

```
# imports y datos
import pandas as pd
df = pd.read_csv('./data/ventas.csv')
df
```

```
# EXPLORO LAS COLUMNAS DEL DF
print(df.columns)

# Iteramos los nombres de las columnas
for c in df.columns:
    print(c)
```

```
# cuantos registros tiene el dataframe?
msj = f'El dataframe tiene {df.index.size} registros.'
print(msj)
```

```
# qué tipo de datos tiene?
df.dtypes
```

```
#CONVIERTO LA COLUMNAS DE importes a FLOAT
df.Importe = df.Importe.astype(float)
df.Comision_x_venta = df.Comision_x_venta.astype(float)
```

```
# VERIFICO
# qué tipo de datos tiene ahora?
df.dtypes
```

```
# LE DOY FORMATO DE $$$$

pd.options.display.float_format = '$ {:,.2f}'.format # Configuramos separadores de miles y 2 decimales
print(df)
```

```
#ORDENAR EL DATAFRAME POR NOMBRE DEL VENDEDOR
print(df.sort_values('Vendedor')) # Muestra una VISTA ordenada del df, pero no lo modifica
df
```

```
#OBTENER EL MÁXIMO DE TODAS LAS COLUMNAS
df.max()
```

```
# ACTIVIDAD DEL VENDEDOR Mufi Panchin con boolean masking
print(df.Vendedor == 'Mufi Panchin')
```

```
# obtengo los registros completos de actividad de Mufi Panchin
v10 = df.loc[ df.Vendedor == 'Mufi Panchin'] # EXITO!!! -> método .loc
print(v10)
```

```
# BUSCO LOS REGISTROS DONDE EL VENDEDOR SEA "Mufi Panchin" y Sede sea "Paternal"
v11 = df.loc[ (df.Vendedor == 'Mufi Panchin') & (df.Sede == 'Paternal')]
print(v11)
```

```
# CARGO LAS MEJORES VENTAS (>200000) EN OTRO DATAFRAME
mejores_ventas = df.loc[  df.Importe > 200000    ]
# mejores_ventas

# Y BUSCO A MUFI PANCHIN en el nuevo df

v10_2 = mejores_ventas.loc[ mejores_ventas.Vendedor == 'Mufi Panchin' ]
v10_2
```

```
# cuántas operaciones >200000 concretó Mufi Panchin?
msj = f'El vendedor Mufi Panchin concretó {v10_2.index.size} operaciones mayores que $200.000'
print(msj)
```

```
# QUIERO VER QUÉ RUBROS VENDIÓ MUFI PANCHIN Y POR QUÉ IMPORTE
print(v10_2[ ['Rubro','Importe'] ]) #Para seleccionar algunas columnas se usa una lista con los nombres
```

```
# si el importe es menor a 350000, cambiarlo a None(ninguno)

v10_2.loc[ v10_2.Importe < 350000, 'Importe' ] = None
print(v10_2[ ['Rubro','Importe'] ])
```

```
# cuando haga un count, Los None van a ser ignorados
print('Total de operaciones de Mufi Panchin por más de $350.000: ', v10_2['Importe'].count())
```

```
# si necesito que no queden valores NaN que sean ignorados por algunas de las funciones numpy, 
# puedo rellenarlos con algún  valor predeterminado. En el ejemplo, con 0
v10_2 = v10_2.fillna(0)
print(v10_2)


```

```
# o pasar lo que necesito a otro df
v10_3 = v10_2.loc[ v10_2.Importe > 0  ]
print(v10_3)
```
