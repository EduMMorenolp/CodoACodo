<h3>Pyplot</h3>
<p>Es la funcionalidad de Matplotlib para mostrar las gráficas</p>

```
# Imports y datos
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 11)
y = x ** 2

print(x)
print(y)
```

<h4>Personalización: Colores</h4>
<small>Documentación: <a href='https://matplotlib.org/3.5.0/gallery/color/named_colors.html'>https://matplotlib.org/3.5.0/gallery/color/named_colors.html</a></small>

```
plt.plot(x,y,'r')
plt.plot(x,y+2,'tab:purple')
plt.plot(x,y+4,'darkturquoise')
plt.plot(x,y+6,'#007700') # Digitos hexadecimales (0 - F) - 00 a FF - 0 a 255
plt.show()
```

<h4>Personalización: Markers</h4>
<small>Documentación: <a href='https://matplotlib.org/3.5.0/api/markers_api.html?highlight=markers#module-matplotlib.markers'>https://matplotlib.org/3.5.0/api/markers_api.html?highlight=markers#module-matplotlib.markers</a></small>
<p>Revisa la lista completa de markers en la documentación.</p>
<p>Algunos de los más utilizados:</p>
<img src ='/work/img/markers.png'>


```
# puedo incluir el marker con el color si uso base colors. Si además quiero la línea, pongo *-

plt.plot(x,y,'r*')
plt.plot(x,y+2,'tab:purple',marker='p')
plt.plot(x,y+4,'darkturquoise', marker='4')
plt.plot(x,y+6,'#007700', marker='s') # Digitos hexadecimales (0 - F) - 00 a FF - 0 a 255
plt.show()
```

<h4>Personalización: Estilos de líneas</h4>
<ul>
<li>Sólida: '-'</li>
<li>Guiones: '--'</li>
<li>Guión y punto: '-.'</li>
<li>Punteado: ':'</li>
<li>Ninguna: None, '', o ' '</li>
</ul>

```
# puedo incluir el marker con el color si uso base colors. Si además quiero la línea, pongo *-
plt.plot(x,y,'r*-')
plt.plot(x,y+2,'tab:purple',marker='p', linestyle=':')
plt.plot(x,y+4,'darkturquoise', marker='4', linestyle='-.')
plt.plot(x,y+6,'#007700', marker='s', linestyle='--') # Digitos hexadecimales (0 - F) - 00 a FF - 0 a 255

plt.savefig('./img/parabolas.png')
plt.show()
```

```
# Puedo utilizar otros tipos de gráfica
print(y)
plt.hist(y,color='tab:purple', width=2)

plt.show()
```

```
# ¿cómo buscar las opciones para cada gráfica en la documentación?
# https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.scatter.html
plt.scatter(x,y,marker='8',c='hotpink')

plt.show()
```

```
# Distribución caja y bigotes
plt.boxplot(x)
plt.show()
```

```
plt.boxplot(y)
plt.show()
```

```
plt.pie(y)
plt.show()
```

```
e = (0,0,0.5,0,0,0,0,0,0,0.3,0) # parámetros explode -> cuál y cuánto separar la porción

plt.pie(y,radius=1.5, explode=e)
plt.show()
```

```
print("LABELS y COLORES")
l = ("Ananá", "Naranja", "Palta", "Tomate")
sizes = [245, 130, 230, 150]
e = (0,0,0.1,0)
c = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes,explode=e, labels=l, colors=c)
plt.show()
```

```
# STARTANGLE
print("STARTANGLE : 'Angulo")

l=("Ananá", "Naranja", "Palta", "Tomate")
sizes = [245, 130, 230, 150]
e =(0,0,0.1,0)
c = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, explode=e, labels=l, colors=c, startangle=90)

plt.show()
```

```
# AUTOPCT
print("AUTOPCT: Mostrar porcentajes")
'''
Este parámetro sigue una sintaxis estándar, para que muestre 1 decimal se usa: "%1.1f%%". 
Esto le dice que vamos a mostrar un numero float con 1 decimal y el %% indica que queremos el simbolo %.
'''
l=("Ananá", "Naranja", "Palta", "Tomate")
sizes = [245, 130, 230, 150]
e =(0,0,0.1,0)
c = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, explode=e, labels=l, colors=c, startangle=45, autopct='%1.2f%%')

plt.show()
```

```
# DONUT CHART
print("DONUT CHART")
l=("Ananá", "Naranja", "Palta", "Tomate")
sizes = [245, 130, 230, 150]
e =(0.1,0.1,0.1,0.1)
c = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.pie(sizes, explode=e, labels=l, colors=c, startangle=45, autopct='%1.2f%%', pctdistance=0.75)

centre_circle = plt.Circle((0,0), 0.6, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```

<h3>Guardando los gráficos</h3>
<p>Todos los gráficos que estuvimos haciendo se pueden guardar como PNG o JPG.</p>
<p>Esto lo hacemos con la siguiente función:</p>
<code>plt.savefig(nombre del archivo)</code>
<p>El archivo se agrega a la lista de Files del proyecto.</p>
<h4>Notas:</h4>
<p>Es muy imporante que se hagan antes de plt.show() ya que matplotlib habrá enviado el gráfico a la pantalla y este se limpia para hacer otro gráfico!</p>
<p>Se puede agregar el parámetro opcional dpi para obtener la resolución deseada (se especifica el número de puntos por pulgada):</p>
<code>plt.savefig('plot.png', dpi=300)</code>
