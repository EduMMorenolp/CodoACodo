# 1.- Códigos de consultas con python

## a) Importar las librerías

<b>i.</b> **Primer método** para la conexión `get_connetion()`
<b>ii.</b> **Segundo método** para la desconexión `close_connetion()`

```
# Importacion de librerias

import sqlite3 as sq3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# CONEXION
database = '/work/mi_proyecto/Consigna#5R/crud/netflix_oscar.db'

def get_connection():
    conn = sq3.connect(database)
    cur = conn.cursor()
    return conn, cur

# DESCONEXION
def close_connection():
    cur.close()
    conn.close()
```

```
# Conectar a la base de datos
conn, cur = get_connection()

# Consulta para seleccionar todos los registros y columnas de la tabla 'content'
query_ver_todo_content = """
SELECT *
FROM content;
"""

query_ver_todo_production = """
SELECT *
FROM production;
"""

# Obtener el resultado en un DataFrame
df_ver_todo_content = pd.read_sql(query_ver_todo_content, conn)

df_ver_todo_production = pd.read_sql(query_ver_todo_production, conn)

# Cerrar la conexión
close_connection()

# Mostrar el DataFrame con la base de datos completa
print("Base de datos completa (tabla 'content'):")
df_ver_todo_content
```

```
print("Base de datos completa (tabla ' production'):")
df_ver_todo_production
```

# 2.- Querys, python y DF de Pandas:

## b)  Los no nulos para la tabla content de la BBDD

```
# Conectar a la base de datos
conn, cur = get_connection()

# Consulta SQL
query = """
SELECT 
    COUNT('type') as 'type',
    COUNT('title_content') as 'title_content',
    COUNT('director') as 'director',
    COUNT('cast') as 'cast',
    COUNT('country') as 'country',
    COUNT('date_added') as 'date_added',
    COUNT('release_year') as 'release_year',
    COUNT('rating') as 'rating',
    COUNT('duration') as 'duration',
    COUNT('listed_in') as 'listed_in',
    COUNT('description') as 'description'
FROM content;
"""

# Ejecutar la consulta y obtener el resultado en un DataFrame
df = pd.read_sql(query, conn)
df = df.rename(index={0: 'No nulos'})

# Mostrar el DataFrame
print("DataFrame:")
display(df.T)

# Cerrar la conexión
close_connection()

```

## c) Realizar cada una de las siguientes instrucciones

<b>i.</b>  Listar todos los registros de la tabla, visualizando las columnas type, title_content, country, rating, duration, listed_in. 
    <b>ii.</b>  Generar un listado de todas las películas en las que participe Argentina, mostrando las columnas id_content, type, title_content, director, cast, rating, duration, listed_in.
    <b>iii.</b>  Presentar un listado con las columnas id_production, title_production, genre, language, imdb_score de los registros de la tabla en las que el imdb_score sea superior a 7.5.
    <b>iv.</b> Seleccionar los títulos con runtime superior a 90 y lenguaje Francés. Incluir el tipo de contenido. Ordenar los títulos de menor a mayor.

```
# Conectar a la base de datos
conn, cur = get_connection()

# i. Listar todos los registros de la tabla 'content'
query_listar_registros = """
SELECT
    type,
    title_content,
    country,
    rating,
    duration,
    listed_in
FROM content;
"""
df_listar_registros = pd.read_sql(query_listar_registros, conn)

# ii. Generar un listado de todas las películas en las que participe Argentina
query_peliculas_argentinas = """
SELECT
    type,
    title_content,
    country,
    rating,
    duration,
    listed_in
FROM content
WHERE country = 'Argentina';
"""
df_peliculas_argentinas = pd.read_sql_query(query_peliculas_argentinas, conn)

# iii. Listado de las columnas id_production, title_production, genre, language, imdb_score
#     de los registros de la tabla en los que el imdb_score sea superior a 7.5 de la tabla 'production'. 
query_altos_imdb_scores = """
SELECT
    title_production,
    genre,
    language,
    imdb_score
FROM production
WHERE imdb_score > 7.5;
"""
df_altos_imdb_scores = pd.read_sql(query_altos_imdb_scores, conn)

# iv. Seleccionar los títulos con runtime superior a 90 y lenguaje Francés.
#Incluir el tipo de contenido. Ordenar los títulos de menor a mayor.
query_titulos_runtime_lenguaje = """
SELECT
    title_production,
    runtime,
    language
FROM production 
WHERE runtime > 90 AND language = 'French'
ORDER BY title_production ASC;
"""

df_titulos_runtime_lenguaje = pd.read_sql(query_titulos_runtime_lenguaje, conn)

# Cerrar la conexión
close_connection()

# Mostrar los DataFrames resultantes
print("i. Listar todos los registros de la tabla 'content':")
df_listar_registros
```

## e)  Código en python que permita realizar las siguientes instrucciones en la BBDD: 
Para ello: 

   <b>i.</b> Realizar un gráfico que muestre la relación entre la cantidad de series y películas disponibles en Netflix.
   <b>ii.</b> Realizar un gráfico a partir de los países que tienen mayor cantidad total de títulos en la plataforma, desglosando por tipo de contenido, mostrando la información sólo para los primeros 10 países (mostrar el desglose por tipo de contenido), Obvie las cooperaciones entre países.

```
# i. Realizar un gráfico que muestre la relación entre la cantidad de series y películas disponibles en Netflix
df_content_types = df_ver_todo_content.groupby('type').size()

# Calcular la suma total
total = df_content_types.sum()

# Crear un gráfico de barras con las barras de 'TV Show', 'Movie' y 'Total'
plt.figure(figsize=(10, 6))
# Barras para 'TV Show'
plt.bar('TV Show', df_content_types['TV Show'], color='red', label='TV Show')
# Barras para 'Movie'
plt.bar('Movie', df_content_types['Movie'], color='blue', label='Movie')
# Barra para 'Total'
plt.bar('Total', total, color='orange', label='Total', alpha=0.5)

plt.title('Relación entre la cantidad de series y películas en Netflix')
plt.xlabel('Tipo de contenido')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```

```
# ii. Realizar un gráfico a partir de los países que tienen mayor cantidad total de títulos en la plataforma, desglosando por tipo de contenido
# Mostrar la información solo para los primeros 10 países
top_countries = df_ver_todo_content['country'].value_counts().head(10).index

# Filtrar el DataFrame para incluir solo los países principales
df_top_countries = df_ver_todo_content[df_ver_todo_content['country'].isin(top_countries)]

# Crear un gráfico de barras apiladas
plt.figure(figsize=(12, 8))
df_top_countries.groupby(['country', 'type']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Top 10 países con mayor cantidad de títulos en Netflix (desglose por tipo de contenido)')
plt.xlabel('País')
plt.ylabel('Cantidad')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de contenido')
plt.show()
```
