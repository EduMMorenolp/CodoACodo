## H. Actualizar


```
def actualizar():
    # Crear una tupla con los datos
    datos = (
        type.get(),
        title_content.get(),
        director.get(),
        country.get(),
        release_year.get(),
        rating.get(),
        duration.get(),
        listed_in.get()
    )

    # Actualizar el registro en la BBDD
    try:
        query = "UPDATE content SET type=?, director=?, country=?, release_year=?, rating=?, duration=?, listed_in=? WHERE title_content=?"
        values = (
            type.get(),
            director.get(),
            country.get(),
            release_year.get(),
            rating.get(),
            duration.get(),
            listed_in.get(),
            title_content.get() 
        )
        cur.execute(query, values)
        conn.commit()
        mensaje_registro(datos, 'actualizado')
        limpiar()
    except sq3.Error as e:
        print(f'Error al actualizar el registro: {e}')
        messagebox.showerror('STATUS', 'Error al actualizar el registro.')
```

## I. Borrar


```
def borrar():
    # Llamar a la función leer_campo
    resultado_query = leer_campo(
        title_content.get(), 'Título', 'title_content')

    # Verificar si el título no existe
    if not resultado_query:
        mensaje_campo_existe_o_no(title_content.get(), 'título', 'no')
    else:
        # Si el título existe, llamar a la función rellenar_campos y obtener el primer registro de la lista
        rellenar_campos(resultado_query)
        primer_registro = resultado_query[0]

        # Eliminar el registro
        try:
            query = "DELETE FROM content WHERE title_content=?"
            cur.execute(query, (title_content.get(),))
            conn.commit()

            # Enviar un mensaje de registro eliminado
            mensaje_registro(primer_registro, 'eliminado')

            # Limpiar los campos
            limpiar()

        except sq3.Error as e:
            print(f'Error al borrar el registro: {e}')
            messagebox.showerror('STATUS', 'Error al borrar el registro.')
```
