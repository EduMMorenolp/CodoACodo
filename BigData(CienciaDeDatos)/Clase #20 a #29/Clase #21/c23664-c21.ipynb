## Funciones BUILT-IN (predefinidas)
Son funciones que ya vienen definidas en la biblioteca estandar de Python:
[Funciones Built-In](https://docs.python.org/es/3/library/functions.html)

Algunos ejemplos son las funciones de creacion de objetos y/o cambio de tipo:

`bool() - dict() - float() - int() - list() - set() - str() - tuple()`

Tambien las funciones de salida/entrada por consola:

`print() - input()`

Funciones de informacion sobre objetos:

`len() - type() - id() - help()`

Funciones que modifican valores:

`abs() - bin() - chr() - hex() - oct() - ord() - pow() - round()`

Funciones relacionadas a objetos iterables:

`all() - any() - enumerate() - max() - min() - range() - reversed() - sorted() - sum() - zip()`

```python
# Muestra la longitud o cantidad de elementos de un iterable
print(len('murcielago'))
print(len([34,44,54]))
```

```python
# Muestra el ID unico de cualquier objeto de Python
l1 = [1,2,3]
l2 = [1,2,3]
print('id(l1) :',id(l1))
print('id(l2) :',id(l2))
print('l1 = l2 : ',l1 == l2)
print('id(l1) = id(l2) :', id(l1) == id(l2))
```

```python
# Ayuda sobre objetos y funciones de la libreria estandar
help(sum)
```

```python
# Informacion sobre iterables

# Todos los elementos son True
print('all([1,1,1,1,1]) =', all([1,1,1,1,1]))
print('all([1,1,0,1,1]) =', all([1,1,0,1,1]))
print()

# Alguno de los elementos son True
print('any([0,0,1,0,0]) =', any([0,0,1,0,0]))
print('any([0,0,0,0,0]) =', any([0,0,0,0,0]))
print()

# Maximo y minimo valor
print('max(150,50,10,100) =', max(150,50,10,100))
print('min(150,50,10,100) =', min(150,50,10,100))
print()

# Inversion, ordenado y suma
print('reversed([150,50,10,100]) =',list(reversed([150,50,10,100])))
print('sorted([150,50,10,100]) =',list(sorted([150,50,10,100])))
print('sum([150,50,10,100]) =',sum([150,50,10,100]))
print()

# Compaginar iterables
letras = ['uno','dos','tres']
numeros = [1,2,3]
print(f'zip({letras}, {numeros}) :', list(zip(letras, numeros)))
print(f'zip({numeros}, {letras}) :', list(zip(numeros, letras)))
```

## Metodos de listas

```python
nombres = ['Ana', 'Carlos', 'Delia', 'Ana', 'Silvia', 'Esteban']
print(nombres)
print()

nombres.append('Xenon') # agrega al final
print(nombres)

nombres.insert(4,'Liliana') # inserta en la posicion
print(nombres)

nombres.pop() # borra al final
print(nombres)

nombres.remove('Delia') # borra un elemento x valor
print(nombres)

del nombres[2] # Elimina un elemento por posicion
print(nombres)


nombres.clear() # vacia la lista
print(nombres)
```

```python
nombres = ['Ana', 'Carlos', 'Delia', 'Ana', 'Silvia', 'Esteban']
nombres.reverse() # Invierte el orden (no ordena)
print(nombres)

nombres.sort() # ordena creciente
print(nombres)
nombres.sort(reverse=True) # ordena decreciente
print(nombres)
```

```python
l1 = [1,2,3]
l2 = l1 # Crear un nuevo ALIAS de l1
print('l1 =',l1)
print('l2 =',l2)
print()

l2[2] = 1000
print('l1 =',l1)
print('l2 =',l2)
print()

print('id(l1):',id(l1))
print('id(l2):',id(l2))
print()

l3 = l1.copy() # Crea una copia independiente de l1
print(l3)
l3[2] = 3 # Modifica el elemento del objeto l3
print()

print('l1 =',l1)
print('l2 =',l2)
print('l3 =',l3)
print()
```

## Metodos de diccionarios

```python
dic1 = {
    'cant':123,
    'tipo':'nuevo',
    'descripcion':'parte de arriba'
}
print(dic1)

dic1['tipo'] = 'usado' # Modifica el valor xq la clave ya existe
print(dic1)

dic1['origen'] = 'Rumania' # Agrega un elemento xq la clase no existe
print(dic1)

dic1.pop('tipo')
print(dic1)

print(dic1.items()) # Obtiene una lista de tuplas (clave,valor)
print(dic1.keys())  # Obtiene una lista de claves
print(dic1.values())  # Obtiene una lista de valores

# Actualiza un diccionario aplicandole los cambios indicados en otro diccionario
dic1.update({'type':'used', 'destino':'Argentina', 'origen':'Noruega'})
print(dic1)
```

```python
# Ejemplo de creacion de un diccionario a partir de 2 listas: 1 de claves y 1 de valores
claves = ['nombre', 'edad','dni']
valores = ['Juan', 32, 45012345]
print('claves:',claves)
print('valores:',valores)

compaginado = list(zip(claves, valores))
print('zip(claves,valores):',compaginado)

persona = dict(compaginado)
print('dict(compaginado):',persona)
```
