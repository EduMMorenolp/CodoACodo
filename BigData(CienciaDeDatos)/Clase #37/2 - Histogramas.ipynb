<h4>Histogramas</h4>
<p>Los gráficos de histograma son una excelente manera de visualizar distribuciones de datos: en un histograma, cada barra agrupa números en rangos. Las barras más altas muestran que más datos caen en ese rango.</p>
<p>Un histograma muestra la forma y la extensión de los datos de muestra continuos.</p>
<p>Para crearlos utilizamos la función</p>
<code>
plt.hist(x, bins, propiedades)
# x = secuencia de datos<br>
# bins = nro de separaciones de mismo ancho<br>
# propiedades: opcional<br>
</code>

<p>Aquí, los contenedores de películas (rangos) se establecen en 10 años. Cada barra aquí incluye todos los programas / películas en lotes de 10 años. Por ejemplo, podemos ver que se lanzaron alrededor de ~ 750 programas entre 2000 y 2010. Al mismo tiempo, ~ 5000 se lanzaron entre 2010. y 2021.</p>

<p>Estos son rangos bastante grandes para la industria del cine, por lo que tiene más sentido visualizarlos para rangos menores a 10 años.</p>

<p>Vamos a extraer la columna DataFrame en una variable data, solo para que sea un poco más fácil de trabajar.</p>

<p>Pasamos data a la función hist(), y configuramos el argumento bins. Acepta una lista, que puede configurar manualmente, especialmente si necesitamos una distribución no uniforme.</p>

<p>Dado que nos gustaría agrupar estas entradas en el mismo período de tiempo (1 año), crearemos una matriz Numpy, que comienza con el valor más bajo (min(data)), termina en el valor más alto (max(data)) y va en incrementos de 1.</p>

Dado que estamos trabajando con intervalos de 1 año, esto dará como resultado la probabilidad de que se estrene un programa en ese año.</p>
