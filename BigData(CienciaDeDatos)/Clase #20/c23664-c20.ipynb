## Bucles (continuacion)

### Desempaquetado de TUPLAS

````python
n = (12,45,8)
num = 1,2,3 # Una tupla tambien se puede crear sin usar parentesis
print(num)

'''
a = num[0]
b = num[1]
c = num[2]
'''

a,b,c = num # Desempaquetado de una tupla
print(a,b,c)

a,b,c,d = 45, 17, 64, -8
print(a,b,c,d)

a,b,c,*d = 45, 17, 64
print(a,b,c,d)

a,b,c,*d = 45, 17, 64, -8, 101, -1000
print(a,b,c,d)
````

### El metodo `.enumerate()`
Este metodo devuelve una lista de tuplas, donde cada tupla contiene el indice y el valor de cada elemento del objeto original.
```python
['Ana','Beto','Carlos'].enumerate()

=> [(0,'Ana'),(1,'Beto'),(3,'Carlos')]
```


````python
lista = [100, 200]
print(list(enumerate(lista)))
````

### Iterar una lista o tupla obteniendo indices y valores

````python
# Iterar una lista (incluyendo indices)
comidas = ['asado', 'pure', 'guiso', 'pizza','guiso']

for i,plato in enumerate(comidas): # Se desempaquetan las tuplas generadas por enumerate()
  print(i,'-',plato.capitalize())
````

### Iterar un diccionario

````python
elefante = {
    'reino':'animal',
    'orden':'mamifero',
    'patas':4,
    'peso':5000
}

for clave in elefante:
  print(clave,':',elefante[clave])
````

### El metodo `.items()`
Este metodo devuelve una lista de tuplas.

Cada tupla contiene la clave y el valor de cada elemento del diccionario.
```python
{
  'Nombre':'Carlos',
  'Apellido':'Sanchez',
  'Edad':35
}.items()

=> [('Nombre','Carlos'),('Apellido','Sanchez'),('Edad',35)]
```

````python
elefante.items()
````

### Iterar un diccionario obteniendo claves y valores

````python
elefante = {
    'reino':'animal',
    'orden':'mamifero',
    'patas':4,
    'peso':5000
}

for clave,valor in elefante.items(): # Se desempaquetan las tuplas generadas por items()
  print(clave,':',valor)
````

### Breakpoints
- break
- continue
- pass


````python
# Uso de BREAK
for n in range(20):
  print(n, end=' ')
  if(n == 12):
    break # Finaliza el bucle
  if(n%3 == 0):
    print('*', end=' ')
````

````python
# Uso de CONTINUE
for n in range(20):
  print(n, end=' ')
  if(n == 12):
    continue # Finaliza la repeticion actual del bucle
  if(n%3 == 0):
    print('*', end=' ')
````

````python
# Uso de PASS
for n in range(20):
  print(n, end=' ')
  if(n == 12):
    pass #Aun no decido que voy a hacer en este caso
  if(n%3 == 0):
    print('*', end=' ')
````

## Funciones
Es una porcion de codigo reutilizable.

Una funcion tiene NOMBRE.

Opcionalmente una funcion tiene PARAMETROS (cualquier cantidad).

Opcionalmente una funcion tiene VALOR DE RETORNO (unico).

### Funcion sin parametros

````python
# Definir una funcion (sin parametros)
def saludar():
  print('¡Hola!')

# Llamar o invocar una funcion
saludar()
a=1
b=2
c=3
saludar()
d=4
saludar()
````

### Funcion con parametros
Los `parametros` de una funcion reciben una copia de los valores de los `argumentos` que se envian durante las llamada a la funcion.

````python
# Definicion de una funcion con parametros
def saludarN(nombre): # las variables entre parentesis se llaman PARAMETROS
  print('¡Hola',nombre+'!')

n = 'Silvia'
saludarN(n)
saludarN('Jose')
saludarN(n + ' Gutierrez') # Cualquier valor que se pasa en la llamada se llama ARGUMENTO
````

### Funcion con valor de retorno
El valor que retorna una funcion se puede:
- asignar a una variable
- usar dentro de una expresion o calculo
- mostrar por pantalla
- usar como argumento de una llamada a otra funcion
- etc.

````python
# Definicion de una funcion con valor de retorno
def promedio(n1,n2,n3):
  p = (n1 + n2 + n3) / 3
  return p # Devuelve este valor al procedimiento que hizo la llamada

prom = promedio(5,7,9)
````
