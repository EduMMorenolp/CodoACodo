```
import numpy as np
import pandas as pd
```

```
# Serie
proveedores = pd.Series(['Apple', 'Microsoft', 'LG', 'Samsung', 'AMD', 'IBM'])
proveedores
```

```
# Serie indexada
proveedores = pd.Series(['Apple', 'Microsoft', 'LG', 'Samsung', 'AMD', 'IBM'], index = ['P-1254', 'P-1285','P-1587','P-1985','P-1258','P-1244'])
proveedores
```

```
# DataFrame indexado
provs = np.array([[None,'Empresa','Total'],['P-1254','Apple',235.5],['P-1285','Microsoft',325.56],['P-1587','LG',125.45],['P-1985','Samsung',548.32],['P-1258','AMD',562.44],['P-1244','IBM',561.88]])
pprovs = pd.DataFrame(data=provs[1:,1:], index=provs[1:,0], columns=provs[0,1:])
print(pprovs)
print()
print(pprovs.dtypes)
print()
pprovs.Total = pprovs.Total.astype(float)
print(pprovs.dtypes)
```

## Diccionarios
<p>Ya vimos que podemos obtener los datos de distintos tipos de archivos.<br>
Veamos cómo trabajar con diccionarios y archivos JSON.</p>

```
# Serie a pártir de un diccionario
dict_prov = {'P-1254':'Apple', 'P-1285':'Microsoft','P-1587':'LG','P-1985':'Samsung','P-1258':'AMD','P-1244':'IBM'}

pv = pd.Series(dict_prov)
pv
```

```
# Diccionario 2
dict_prov2 = {'Empresa':['Apple', 'Microsoft', 'LG', 'Samsung', 'AMD', 'IBM'],
             'id': ['P-1254', 'P-1285','P-1587','P-1985','P-1258','P-1244'],
             'artículos':[['computadoras','celulares'],['software'],['monitores','celulares'],['monitores','celulares','electrodomésticos'],['procesadores'],['software','computadoras']]}

pv2 = pd.DataFrame(dict_prov2)

pv2.set_index('id', inplace=True)

# pv2 = pv2.set_index('id')

pv2
```

```
# JSON a dataframe
hp = pd.read_json('./data/hpcharactersdataraw.json')
hp
```

```
# Cantidad de registros
print(hp.index.size)
print()

# Nombres de las columnas
print(hp.columns)
print()

# Obtener todos datos de un registro
print(hp.loc[4])
print()
print(hp.iloc[4])
print()

# Obtener un dato específico
print(hp.loc[4,['Blood','Profession']])
```

```
# Un JSON también se puede levantar como un serie de diccionarios

hp_series = pd.read_json('./data/hpcharactersdataraw.json', typ='Series')
hp_series
```

```
# SLICING
print(hp[0:11])
```

```
# Filtrado con Slicing
print(hp.loc[0:5,['Name','Blood'] ]) # Slice con .loc alcanza el indice tope
```

```
print(hp.loc[ hp.Name.str[0] == 'A' ,['Name','Blood'] ])
```

<h3>Filtrado por loc/iloc</h3>

```
# ¿Quiénes son Muggles?
#muggles = hp[ (hp.Blood == 'Muggle-born')]
muggles = hp[ (hp.Blood == 'Muggle-born') | (hp.Blood == 'Muggle')]
print(muggles.Name)
print(muggles.Name.size)
```
