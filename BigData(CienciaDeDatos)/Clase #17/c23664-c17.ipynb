## Funcion type()

```python
print(type(8))
print(type(5.7))
print(type('mi casa'))
print(type(True))
print()
print(type([12, 4.5]))
print(type((12, 4.5)))
print(type({12, 4.5}))
print(type({'peso':12, 'altura':4.5}))
print()
print(type(range(13)))
```
