## Metodos de los Strings

### Inmutabilidad de los objetos String

```python
frase = "La comision 23553 tiene 33 alumnos"
print(frase[12])

# frase[12] = '3' # Da error xq los strings son inmutables

s = 'casa'
print(id(s))
s = 'perro'
print(id(s)) # este objeto "s" es diferente al anterior "s" (diferente id)
```

### Mayusculas y minusculas

```python
p = 'murCieLAgo bLAncO'

print(p.capitalize())
print(p.upper())
print(p.lower())
print(p.swapcase())
print(p.title())
```

### Formateo de numeros

```python
n = '365'

print(n)
print(n.zfill(5))
print('3'.zfill(5))
```

### Manejo de espacios iniciales y finales

```python
p = "   algo cualquiera   "
print(p.lstrip())
print(p.rstrip())
print(p.strip())
```

### Busquedas

```python
s = "caballo blanco"
print(s.find('llo b'))
print(s.count('a'))
print('llo b' in s)
```

### Comprobaciones
Son metodos que devuelven valores bool.

```python
s = 'A1'
print(s.isalnum())

s = 'Ab'
print(s.isalpha())

s = '34'
print(s.isnumeric())
print()

print('AAAA'.isupper())
print('AAaA'.isupper())
print()
print('eeee'.islower())
print('eeEe'.islower())
```

### Comienza , termina o contiene

```python
print('Alpaca'.startswith('Al'))  # Comienza con
print('Alpaca'.endswith('ca'))    # Termina con
print('lpa' in 'Alpaca')          # Contiene a
```

### Conversiones String <=> Lista

```python
# Conversiones de String a Lista

s = "La casita de la abuela de Caperucita Roja"
l = s.split()
print(l)

s = "Juan,Perez,45,Marketing"
l = s.split(',')
print(l)
```

```python
# Conversiones de Lista en String
l = ['Juan','Perez',45,'Marketing']

l1=[]
for e in l:
  l1.append(str(e))
print(l1)

s = ','.join(l1)  # Junta todos los elementos, separandolos con una ","
print(s)
```

### Reemplazo de substrings

```python
frase = "Me gusta el helado de pistacho"
s = frase.replace('pistacho','chocolate')
print(frase)
print(s)

csv = '23,Juan,45,NULL'
print(csv)
csv = csv.replace('NULL','None')
print(csv)
```

## Listas de comprension
Son listas generadas mediante un algoritmo.

`lista = [ algoritmo de generacion de elementos ]`

```python
# Listas inteligente (de comprension)
# lista = [5,6,7,8,9,10,11]
lista = list(range(5,12,2))
print(lista)

# Quiero los años bisiestos desde 1600 a 2023
# Los años biciestos son los divisibles x 4,
# pero que no sean divisibles x 100 (con la excepcion de los divisibles x 400)
bisiestos = [a for a in range(1600,2024) if (a%400==0) or ((a%4==0) and not(a%100==0))]
print()
print('Años bisiestos')
print(bisiestos)
decadas = [a for a in bisiestos if a%10==0]
print(decadas)
```

## SLICES (cortado o trozado)
Es una sintaxis que permite "extraer" un determinado conjunto de elementos de un objeto iterable.

```python
lista = [1600, 1620, 1640, 1660, 1680, 1720, 1740, 1760, 1780, 1820, 1840, 1860, 1880, 1920, 1940, 1960, 1980, 2000, 2020]

print(lista[2])
print(lista[3:6])
print(lista[:6])
print(lista[-3:])
print(lista[:])
print(lista[::3])
print(lista[::-1])
```
