## Modulos (librerias)
- Importacion completa
- Importacion parcial
- Importacion con alias

```python
# Importar un modulo completo
import math

p = math.pi
print(p)
e = math.e
print(e)
print()
print(math.log(e))
print(math.floor(7.9))
print(math.ceil(7.1))
```

```python
# Importar algunos objetos de un modulo
from math import floor, ceil
print(floor(88.6))
print(ceil(88.6))
```

```python
# Importar todos los objetos de un modulo
from math import *
print(sqrt(100))
```

```python
# Importar un modulo con alias
import math as m
print(m.trunc(5.6))
```

```python
# Importar objetos con alias
from math import sqrt as raiz_cuadrada
print(raiz_cuadrada(144))
```

```python
# Ejemplo de uso del modulo matplotlib
# (modulo externo no incluido en el paquete original de Python)
from matplotlib import pyplot as plt
x = [n for n in range(-4,5)]
print(x)
y = [n**2 for n in x]
print(y)
plt.plot(x,y)
plt.show()
```
