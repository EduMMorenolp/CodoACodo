## Operadores logicos
- and (conjuncion) - interseccion (Y)
- or (disyuncion) - union (O)
- not (negacion) - complemento (NO)


````python
print('NOT (negacion)')
print('NOT True =', not True)
print('NOT False =', not False)
print('NOT (9 > 7) =', not (9>7))
print()

print('AND (conjuncion)')
print('False AND False =', False and False)
print('False AND True =', False and True)
print('True AND False =', True and False)
print('True AND True =', True and True)
print()

print('OR (disyuncion)')
print('False OR False =', False or False)
print('False OR True =', False or True)
print('True OR False =', True or False)
print('True OR True =', True or True)
print()
````

## Operadores de pertenencia

## Condicionales (estructura de control)
- Condiciones logicas
- Condicional simple
- Indentacion OBLIGATORIA
- Condicional completo
- Condicionales en cascada
- Regla del N-1 y la salida por defecto (x descarte)

Bloque de instrucciones
- En otros lenguajes los bloques se encierran entre `{ }`

```js
{
  instruccion 1
  instruccion 2
  instruccion 3
}
```
- En Python un bloque se encierra entre `:` y la finalizacion del indentado

```python
estructura_de_control :
  instruccion 1
  instruccion 2
  instruccion 3
```


````python
edad = 18

# Condicional SIMPLE
if (edad >= 18):
  print('Es mayor de edad')
print()
print('---')
````

````python
edad = 17

# Condicional COMPLETO
if (edad >= 18):
  print('Es mayor de edad')
  print('Puede ingresar')
else:
  print('Es menor de edad')
  print('No puede ingresar')
print()
print('---')
````

````python
# Condicionales anidados o en cascada
'''
Supongamos que tenemos una maquina expendedora de bebidas calientes:
1 - Cafe
2 - Te
3 - Chocolate
4 - Mate cocido
'''
print('Expendio de bebidas caliente')
opcion = int(input('Seleccione su opcion: '))

if (opcion == 1):
  print('Sirvase su CAFE.')
else:
  if (opcion == 2):
    print('Sirvase su TE.')
  else:
    if (opcion == 3):
      print('Sirvase su CHOCOLATE.')
    else:
      if (opcion == 4):
        print('Sirvase su MATE COCIDO.')
      else:
        print('Opcion incorrecta') # Esta opcion se llama DEFAULT o DESCARTE
````

````python
# Condicionales anidados o en cascada (mejorado con ELIF)
'''
Supongamos que tenemos una maquina expendedora de bebidas calientes:
1 - Cafe
2 - Te
3 - Chocolate
4 - Mate cocido
'''
print('Expendio de bebidas caliente')
opcion = int(input('Seleccione su opcion: '))

if (opcion == 1):
  print('Sirvase su CAFE.')
elif (opcion == 2):
  print('Sirvase su TE.')
elif (opcion == 3):
  print('Sirvase su CHOCOLATE.')
elif (opcion == 4):
  print('Sirvase su MATE COCIDO.')
else:
  print('Opcion incorrecta') # Esta opcion se llama DEFAULT o DESCARTE
````

## Bucles (estructura de control iterativas)
- Bucle condicional (while)
 - Condicion
 - Actualizar para evitar bucles infinitos
 - Opcional ELSE
- Bucle fijo (for)
 - Iteracion de rangos
 - Iteracion de strings
 - Iteracion de listas (o tuplas)
 - Uso de enumerate
 - Iteracion de diccionarios

````python
# Bucle condicional WHILE
cant_pelotas = 7
while (cant_pelotas > 0):
  print('Sacamos la pelota',cant_pelotas)
  cant_pelotas -= 1
print('---')
````

````python
# Bucle fijo FOR
# Itera (o recorre) objetos iterables
for n in range(5):
  print(n)
````

````python
# Iterar un String
palabra = 'Python'

for letra in palabra:
  print(letra.upper())
````

````python
# Iterar una lista
comidas = ['asado', 'pure', 'guiso', 'pizza']

for plato in comidas:
  print(plato.capitalize())
````
