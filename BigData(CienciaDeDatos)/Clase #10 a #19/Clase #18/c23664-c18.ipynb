## Salidas y entradas de datos por consola

- Salidas de datos: funcion `print()`
 - Separador de datos
 - Terminador de linea
- Entradas de datos: funcion `input()`
 - Mensaje de aclaracion
 - Conversion de tipo del dato ingresado


```python
print(100)
print('hola')
print(False) # El finalizador x defecto es un "fin de linea"
print()

print('La edad de Juan es', 45,'años.') # El separador x defecto es 'un espacio'

# Modificar el separador
print(5,4,3,2,1, sep=' - ')
print(5,4,3,2,1, sep=' * ')
print(5,4,3,2,1, sep='\t')
print()
print(5,4,3,2,1, sep='\n')
print()

# Modificar el finalizador
print('Hola', end=' ')
print('gente')
print('Buenas tardes!')
```

```python
# input() -> no se usa de esta manera xq es confuso para el usuario
input('Ingrese su edad: ')
```

## Operadores aritmeticos
- Suma
- Resto
- Producto
- Cociente
- Potenciacion (y radicacion)
- Cociente entero
- Resto


```python
print('La suma de 4 + 5 es', 4 + 5 )
print('La resta de 4 - 5 es', 4 - 5 )
print('El producto de 4 x 5 es', 4 * 5 )
print('El cociente de 4 / 5 es', 4 / 5 )
print()
print('La potencia de 4 elevado a la 5º es', 4 ** 5 )
print('La raiz 4º de 5 es', 5 ** (1/4) )
print()
print('El cociente entero entre 14 y 3 es', 14 // 3)
print('El resto del cociente entero entre 14 y 3 es', 14 % 3)
print()
print('¿435 es multiplo de 5? :', 435 % 5)
```

## Operadores de cadenas
- Concatenacion
- Repeticion

```python
frase = 'La casa de mi vecino'
print(frase + ' es de color verde.')

palabra = 'Ja '
print(palabra * 5)
```

## Operadores de asignacion
- Suma acumulada
- Resta acumulada
- Producto acumulado
- Cociente acumulado
- Potenciacion acumulada
- Cociente entero acumulado
- Resto acumulado


```python
total_cuenta_pesos = 107000
deposito = 3700
total_cuenta_pesos = total_cuenta_pesos + deposito
print(total_cuenta_pesos)

deposito = 5100
total_cuenta_pesos += deposito
print(total_cuenta_pesos)

edad = 50
edad += 1 # Contador o Incrementador
print(edad)
```

## Operadores relacionales (o de comparacion)
- Menor
- Menor o Igual
- Mayor
- Mayor o Igual
- Igual
- Distinto


```python
# El resultado de una operacion relacional SIEMPRE es un bool
print('8 < 17 =', 8 < 17)
print('8 < 5 =', 8 < 5)
print()
print('8 <= 8 =', 8 <= 8)
print('8 <= 7 =', 8 <= 7)
print()
print('8 > 17 =', 8 > 17)
print('8 > 5 =', 8 > 5)
print()
print('8 >= 8 =', 8 >= 8)
print('8 >= 9 =', 8 >= 9)
print()
print('8 = 8 =', 8 == 8)
print('8 = 9 =', 8 == 9)
print()
print('8 <> 8 =', 8 != 8)
print('8 <> 9 =', 8 != 9)
print()
```
